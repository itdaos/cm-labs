```python
class ES:
    """
    Equation Solver - static object which can be assigned a function
    """
    func: callable    # function - used to represent math equation f(x) = 0 to be solved
    dFunc: callable   # function - used in newtons method = derivative 
    EPS = 10**-6      # epsilon  - calculation precision
    
    @staticmethod
    def sgn(x: int) -> int:
        if (x > 0):
            return 1
        if (x < 0):
            return -1
        return 0

    @staticmethod
    def bisection(a, b):
        print("**Bisection**")
        counter = 0
        if a > b:
            a, b = b, a
        
        while (b-a > ES.EPS):
            counter += 1
            c = (a + b)/2
            if ES.sgn(f(a)) != ES.sgn(f(c)):
                b = c
            elif ES.sgn(f(b)) != ES.sgn(f(c)):
                a = c
            else:
                print("Error: Couldn't find a proper range")
            
            print("\ta = %.6f; b = %.6f; f(a) = %.6f; f(b) = %.6f"  % (a, b, ES.func(a), ES.func(b)))
            
        print("Final result: %.6f\nNumber of iterations: %d" % (c, counter))
        return c
    
    @staticmethod
    def horde(a, b):
        print("**Hordes**")
        counter = 0
        if a > b:
            a, b = b, a
            
        while (b-a > ES.EPS):
            counter += 1
            c = (a*ES.func(b) - b*ES.func(a))/(ES.func(b) - ES.func(a))
            
            if ES.sgn(f(a)) != ES.sgn(f(c)):
                b = c
            elif ES.sgn(f(b)) != ES.sgn(f(c)):
                a = c
            else:
                print("Error: Couldn't find a proper range")
            
            print("\ta = %.6f; b = %.6f; f(a) = %.6f; f(b) = %.6f"  % (a, b, ES.func(a), ES.func(b)))
            
        print("Final result: %.6f\nNumber of iterations: %d" % (c, counter))
        return c
    
    @staticmethod
    def newton(a, b):
        print("**Newtons**")
        counter = 0
        if a > b:
            a, b = b, a
        
       
        if (ES.func(a)*ES.dFunc(a) > 0):
             x: int = a
        else:
             x: int = b
                    
        while (abs(ES.func(x)) > ES.EPS):
            counter += 1
            l = ES.func(x)/ES.dFunc(x)
            x = x - l
            
            print("\tx = %.6f; f(x) = %.6f"  % (x, ES.func(x)))
            
        print("Final result: %.6f\nNumber of iterations: %d" % (x, counter))
        return x
            

def f(x):
    return -2*x**4 + x**3 + 5*x**2 - 2*x + 6

def df(x):
    return -8*x**3 + 3*x**2 + 10*x - 2

ES.func = f
ES.dFunc = df
```

```python
x1 = ES.bisection(-2, -1)
x1_wolf = -1.77085478546621
delta = -abs(x1 - x1_wolf)*100/x1_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))

x2 = ES.bisection(1, 2)
x2_wolf =  1.93696158337786
delta = abs(x2 - x2_wolf)*100/x2_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))
```

```python
x1 = ES.horde(-2, -1)
x1_wolf = -1.77085478546621
delta = -abs(x1 - x1_wolf)*100/x1_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))

x2 = ES.horde(1, 2)
x2_wolf =  1.93696158337786
delta = abs(x2 - x2_wolf)*100/x2_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))
```

```python
x1 = ES.newton(-2, -1)
x1_wolf =  1.93696158337786
delta = abs(x1 - x1_wolf)*100/x1_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))

x2 = ES.newton( 1,  2)
x2_wolf = -1.77085478546621
delta = -abs(x2 - x2_wolf)*100/x2_wolf
print("Relative diversion: %.6f %s" % (delta, "%"))
```

```python
import numpy as np

eps = 1e-5
alpha, betta = 0.658, -1.311
a, b, c, d = 0.233, -1.838, -1.287, 0.504


def F1(x_k):
    x, y = x_k[0], x_k[1]

    x_next = d - np.sin(y + betta)
    y_next = (c - np.cos(x + alpha)) / b

    return np.array([x_next, y_next])


def F2(x_k):
    x, y = x_k[0], x_k[1]

    f1 = np.sin(x + y) + c*x - d
    f2 = x ** 2 + y ** 2 - 1

    return np.array([f1, f2])


def inv_jacobian(x_k):
    x, y = x_k[0][0], x_k[1][0]

    f1dx = np.cos(x + y) + c
    f2dx = 2*x
    f1dy = np.cos(x + y)
    f2dy =  2*y

    w = np.array([[f1dx, f1dy], [f2dx, f2dy]])

    return np.linalg.inv(w)


def newton(x_prev=np.zeros((2, 2))):
    print("NEWTON`s method")
    done = False
    iter_num = 0

    while not done:
        F = F2(x_prev)
        F_norm = np.linalg.norm(F)
        print("Iteration {}, solution:({:.6}, {:.6}), err:({:.6}, {:.6}), err norm:{:.6}".format(
            iter_num,
            x_prev[0][0], x_prev[1][0],
            F[0][0], F[1][0],
            F_norm
        ))

        x_next = x_prev - inv_jacobian(x_prev).dot(F)

        if F_norm < eps and np.linalg.norm(x_next - x_prev) < eps:
            done = True
        x_prev = x_next
        iter_num += 1

    return x_prev


def simple_iter(x_prev=np.zeros((2, 1))):
    print("Simple iteration method")
    done = False
    iter_num = 0

    while not done:

        F = F1(x_prev) - x_prev
        F_norm = np.linalg.norm(F)

        print("Iteration {}, solution:({:.6}, {:.6}), err:({:.6}, {:.6}), err norm:{:.6}".format(
            iter_num,
            x_prev[0][0], x_prev[1][0],
            F[0][0], F[1][0],
            F_norm
        ))

        x_next = F1(x_prev)
        if F_norm < eps and np.linalg.norm(x_next - x_prev) < eps:
            done = True
        x_prev = x_next
        iter_num += 1

    return x_prev


p1 = np.array([[1.3], [1.5]])
p2 = np.array([[-0.9], [-1.1]])
p3 = np.array([[0.3], [0.4]])

simple_iter(p1)
# simple_iter(p2)
newton(p2)
newton(p3)
```
